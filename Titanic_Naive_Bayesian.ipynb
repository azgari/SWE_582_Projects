```python
import pandas as pd
import scipy.stats  as stats
import numpy as np
import math

df_train = pd.read_csv('train.csv',header=0)
df_test  = pd.read_csv('test.csv',header=0)

#df_train['New_Cabin'] = df_train['Cabin'].astype(str).str[0]
#df_test['New_Cabin']  = df_test['Cabin'].astype(str).str[0]

#print("PassengerId: ", df.PassengerId.unique())
#print("Survived: ",    df.Survived.unique())#Discrete
#print("Pclass: ",      df.Pclass.unique())#Discrete
#print("Name: ",        df.Name.unique())#Won't use
#print("Sex: ",         df.Sex.unique())#Discrete
#print("Age: ",         df.Age.unique())#Gaussian
#print("SibSp: ",       df.SibSp.unique())#Discrete
#print("Parch: ",       df.Parch.unique())#Discrete
#print("Ticket: ",      df.Ticket.unique())#????#Won't use
#print("Fare: ",        df.Fare.unique())#Gaussian
#print("Cabin: ",       df.Cabin.unique())#Discrete #Try Binary
#print("Embarked: ",    df.Embarked.unique())#Discrete


df_train_survived = df_train.loc[df_train["Survived"] == 1]
df_train_dead     = df_train.loc[df_train["Survived"] == 0]

total_dead     = len(df_train_dead)
total_survived = len(df_train_survived)

mean_age_survived = df_train_survived['Age'].mean()
std_age_survived  = df_train_survived['Age'].std()

mean_age_dead = df_train_dead['Age'].mean()
std_age_dead  = df_train_dead['Age'].std()

mean_fare_survived = df_train_survived['Fare'].mean()
std_fare_survived  = df_train_survived['Fare'].std()

mean_fare_dead = df_train_dead['Fare'].mean()
std_fare_dead  = df_train_dead['Fare'].std()

print("Test: ")
#print(df_test)

result = np.zeros([len(df_test) + 1, 2], dtype = object)

result[0, 0] = "PassengerId"
result[0, 1] = "Survived"

for index, row in df_test.iterrows():
    #if(index < 5000):
    print("*******************")
    print("i: ", index)
    #Survival Probability(PClass)
    f_pclass_survived = [item for item in zip(df_train_survived["Pclass"], df_train_survived["Survived"])
                    if item[0] == row["Pclass"]]
    
    total_pclass_survived = len(f_pclass_survived)        
    #print("Pclass " + str(row["Pclass"]), "survived ratio", "", total_pclass_survived, "/",
    #      total_survived, "=", str(total_pclass_survived / total_survived))        
    p_pclass_survived = total_pclass_survived / total_survived    
    
    #Death Probability(PClass)   
    f_pclass_dead = [item for item in zip(df_train_dead["Pclass"], df_train_dead["Survived"])
                if item[0] == row["Pclass"]]
    
    total_pclass_dead = len(f_pclass_dead)        
    #print("Pclass " + str(row["Pclass"]), "dead ratio", "", total_pclass_dead, "/",
    #      total_dead, "=", str(total_pclass_dead / total_dead))        
    p_pclass_dead = total_pclass_dead / total_dead
    
    
    #Survival Probability(Sex)
    f_sex_survived = [item for item in zip(df_train_survived["Sex"], df_train_survived["Survived"])
                    if item[0] == row["Sex"]]
    
    total_sex_survived = len(f_sex_survived)        
    #print("Sex " + str(row["Sex"]), "survived ratio", "", total_sex_survived, "/",
    #      total_survived, "=", str(total_sex_survived / total_survived))      
    p_sex_survived = total_sex_survived / total_survived
    
    #Death Probability(Sex)   
    f_sex_dead = [item for item in zip(df_train_dead["Sex"], df_train_dead["Survived"])
                if item[0] == row["Sex"]]
    
    total_sex_dead = len(f_sex_dead)        
    #print("Sex " + str(row["Sex"]), "dead ratio", "", total_sex_dead, "/",
    #      total_dead, "=", str(total_sex_dead / total_dead))  
    p_sex_dead = total_sex_dead / total_dead
    
    if np.isnan(row["Age"]):
        p_Age_survived = 0.5
        p_Age_dead     = 0.5
    else:         
        #Survival Probability(Age)    
        p_Age_survived = stats.norm.pdf(x= row["Age"],  loc = mean_age_survived,  scale= std_age_survived)
        #print("Age " + str(row["Age"]), "survived ratio", "", p_Age_survived)
        
        #Death Probability(Age)   
        p_Age_dead = stats.norm.pdf(x= row["Age"],  loc = mean_age_dead,  scale= std_age_dead)
        #print("Age " + str(row["Age"]), "dead ratio", "", p_Age_dead)
    
    
    #Survival Probability(SibSp)
    f_SibSp_survived = [item for item in zip(df_train_survived["SibSp"], df_train_survived["Survived"])
                    if item[0] == row["SibSp"]]
    
    total_SibSp_survived = len(f_SibSp_survived)        
    #print("SibSp " + str(row["SibSp"]), "survived ratio", "", total_SibSp_survived, "/",
    #      total_survived, "=", str(total_SibSp_survived / total_survived))      
    p_SibSp_survived = total_SibSp_survived / total_survived
    
    #Death Probability(SibSp)   
    f_SibSp_dead = [item for item in zip(df_train_dead["SibSp"], df_train_dead["Survived"])
                if item[0] == row["SibSp"]]
    
    total_SibSp_dead = len(f_SibSp_dead)        
    #print("SibSp " + str(row["SibSp"]), "dead ratio", "", total_SibSp_dead, "/",
    #      total_dead, "=", str(total_SibSp_dead / total_dead))  
    p_SibSp_dead = total_SibSp_dead / total_dead
    
    
    #Survival Probability(Parch)
    f_Parch_survived = [item for item in zip(df_train_survived["Parch"], df_train_survived["Survived"])
                    if item[0] == row["Parch"]]
    
    total_Parch_survived = len(f_Parch_survived)        
    #print("Parch " + str(row["Parch"]), "survived ratio", "", total_Parch_survived, "/",
    #      total_survived, "=", str(total_Parch_survived / total_survived))      
    p_Parch_survived = total_Parch_survived / total_survived
    
    #Death Probability(Parch)   
    f_Parch_dead = [item for item in zip(df_train_dead["Parch"], df_train_dead["Survived"])
                if item[0] == row["Parch"]]
    
    total_Parch_dead = len(f_Parch_dead)        
    #print("Parch " + str(row["Parch"]), "dead ratio", "", total_Parch_dead, "/",
    #      total_dead, "=", str(total_Parch_dead / total_dead))  
    p_Parch_dead = total_Parch_dead / total_dead
    
    
    #Survival Probability(Fare)    
    p_Fare_survived = stats.norm.pdf(x= row["Fare"],  loc = mean_fare_survived,  scale= std_fare_survived)
    #print("Fare " + str(row["Fare"]), "survived ratio", "", p_Age_survived)
    
    #Death Probability(Fare)   
    p_Fare_dead = stats.norm.pdf(x= row["Fare"],  loc = mean_fare_dead,  scale= std_fare_dead)
    #print("Fare " + str(row["Fare"]), "dead ratio", "", p_Age_dead)
    
    
    #Survival Probability(Cabin)
    #f_Cabin_survived = [item for item in zip(df_train_survived["Cabin"], df_train_survived["Survived"])
    #                if item[0] == row["Cabin"]]
    #
    #total_Cabin_survived = len(f_Cabin_survived)        
    ##print("Cabin " + str(row["Cabin"]), "survived ratio", "", total_Cabin_survived, "/",
    ##      total_survived, "=", str(total_Cabin_survived / total_survived))      
    #p_Cabin_survived = total_Cabin_survived / total_survived
    #
    ##Death Probability(Cabin)   
    #f_Cabin_dead = [item for item in zip(df_train_dead["Cabin"], df_train_dead["Survived"])
    #            if item[0] == row["Cabin"]]
    #
    #total_Cabin_dead = len(f_Cabin_dead)        
    #print("Cabin " + str(row["Cabin"]), "dead ratio", "", total_Cabin_dead, "/",
    #      total_dead, "=", str(total_Cabin_dead / total_dead))  
    #p_Cabin_dead = total_Cabin_dead / total_dead  
    
    
    #if pd.isnull(row["Embarked"]):
    #p_Embarked_survived = 0.5
    #p_Embarked_dead     = 0.5
    #print("null", row["Embarked"])
    #else:         
    #Survival Probability(Age)
    
    #Survival Probability(Embarked)
    #print(row["Embarked"])
    f_Embarked_survived = [item for item in zip(df_train_survived["Embarked"], df_train_survived["Survived"])
                    if item[0] == row["Embarked"]]
    
    total_Embarked_survived = len(f_Embarked_survived)        
    #print("Embarked " + str(row["Embarked"]), "survived ratio", "", total_Embarked_survived, "/",
    #      total_survived, "=", str(total_Embarked_survived / total_survived))      
    p_Embarked_survived = total_Embarked_survived / total_survived
    
    #Death Probability(Embarked)   
    f_Embarked_dead = [item for item in zip(df_train_dead["Embarked"], df_train_dead["Survived"])
                if item[0] == row["Embarked"]]
    
    total_Embarked_dead = len(f_Embarked_dead)        
    #print("Embarked " + str(row["Embarked"]), "dead ratio", "", total_Embarked_dead, "/",
    #      total_dead, "=", str(total_Embarked_dead / total_dead))  
    p_Embarked_dead = total_Embarked_dead / total_dead
    
    p_result_dead = 0
    p_result_survived = 0
    
    p_result_dead = p_pclass_dead * p_sex_dead * p_Age_dead * p_SibSp_dead * p_Parch_dead * p_Fare_dead * p_Embarked_dead * total_dead / len(df_train)
    
    p_result_survived = p_pclass_survived * p_sex_survived * p_Age_survived * p_SibSp_survived * p_Parch_survived * p_Fare_survived * p_Embarked_survived * total_survived / len(df_train)
    
    print("dead", p_result_dead)
    print("survived", p_result_survived)
    
    result[index + 1, 0] = row["PassengerId"]
    if p_result_dead > p_result_survived:
        result[index + 1, 1] = 0
    else:
        result[index + 1, 1] = 1
        
print("x")
        
```
